```python
import pandas as pd
from pathlib import Path
import requests
import json
```

```python
natural_state_path = Path(r"Resources/COnsumption by diff Sectors/By STate/Natural Gas Consumption.xlsx").absolute()
landfillgas_state_path = Path(r"Resources/COnsumption by diff Sectors/By STate/Landfill Gas Consumption.xlsx").absolute()
petroleumcoke_state_path = Path(r"Resources/COnsumption by diff Sectors/By STate/Petroleum Coke Consumption.xlsx").absolute()
petroleumliquids_state_path = Path(r"Resources/COnsumption by diff Sectors/By STate/Petroleum Liquids Consumption.xlsx").absolute()
biogenicmunic_state_path = Path(r"Resources/COnsumption by diff Sectors/By STate/Biogenic Municipal Solid Waste Consumption.xlsx").absolute()
wood_state_path = Path(r"Resources/COnsumption by diff Sectors/By STate/WoodWood Waste Biomass Consumption.xlsx").absolute()
coal_state_path = Path(r"Resources/COnsumption by diff Sectors/By STate/Coal Consumption.xlsx").absolute()
```

```python
natural_state_df = pd.read_excel(natural_state_path, header=None)
landfillgas_state_df = pd.read_excel(landfillgas_state_path, header=None)
petroleumcoke_state_df = pd.read_excel(petroleumcoke_state_path, header=None)
petroleumliquids_state_df = pd.read_excel(petroleumliquids_state_path, header=None)
biogenicmunic_state_df = pd.read_excel(biogenicmunic_state_path, header=None)
wood_state_df = pd.read_excel(wood_state_path, header=None)
coal_state_df = pd.read_excel(coal_state_path, header=None)
```

```python
natural_state_df_2023 = natural_state_df.drop(columns=[2, 3], index=[0, 1, 2, 3])
```

```python
natural_state_df_2023.columns = ["State", "Natural Gas(in Million Cubic Feet)"]
natural_state_df_2023 = natural_state_df_2023.set_index("State")
```

```python
natural_state_df_2023
```

```python
landfillgas_state_df_2023 = landfillgas_state_df.drop(columns=[2, 3], index=[0, 1, 2, 3, 4])
landfillgas_state_df_2023.columns = ["State", "Landfill Gas(in Million Cubic Feet)"]
landfillgas_state_df_2023 = landfillgas_state_df_2023.set_index("State")
```

```python
landfillgas_state_df_2023
```

```python
petroleumcoke_state_df_2023 = petroleumcoke_state_df.drop(columns=[2, 3], index=[0, 1, 2, 3, 4])
petroleumcoke_state_df_2023.columns = ["State", "Petroleum Coke(in Thousand Tons)"]
petroleumcoke_state_df_2023 = petroleumcoke_state_df_2023.set_index("State")
```

```python
petroleumcoke_state_df_2023
```

```python
petroleumliquids_state_df_2023 = petroleumliquids_state_df.drop(columns=[2, 3], index=[0, 1, 2, 3, 4])
petroleumliquids_state_df_2023.columns = ["State", "Petroleum Coke(in Thousand Barrels)"]
petroleumliquids_state_df_2023 = petroleumliquids_state_df_2023.set_index("State")

petroleumliquids_state_df_2023
```

```python
biogenicmunic_state_df_2023 = biogenicmunic_state_df.drop(columns=[2, 3], index=[0, 1, 2, 3, 4])
biogenicmunic_state_df_2023.columns = ["State", "Biogenic Municipal Waste(in Thousand tons)"]
biogenicmunic_state_df_2023 = biogenicmunic_state_df_2023.set_index("State")

biogenicmunic_state_df_2023
```

```python
wood_state_df_2023 = wood_state_df.drop(columns=[2, 3], index=[0, 1, 2, 3, 4])
wood_state_df_2023.columns = ["State", "Wood(in Billion Btus)"]
wood_state_df_2023 = wood_state_df_2023.set_index("State")

wood_state_df_2023
```

```python
coal_state_df_2023 = coal_state_df.drop(columns=[2, 3], index=[0, 1, 2, 3, 4])
coal_state_df_2023.columns = ["State", "Coal(in Thousand Tons)"]
coal_state_df_2023 = coal_state_df_2023.set_index("State")

coal_state_df_2023
```

```python
combined_state_df_2023 = pd.concat([
    natural_state_df_2023, coal_state_df_2023, landfillgas_state_df_2023, wood_state_df_2023, petroleumcoke_state_df_2023, petroleumliquids_state_df_2023, biogenicmunic_state_df_2023], axis=1, join="inner"
    )

combined_state_df_2023.to_csv("Resources/COnsumption by diff Sectors/By STate/Combined 2023.csv")
```
